```python
import cv2
from ultralytics import YOLO
import matplotlib.pyplot as plt

# Load the trained model
model = YOLO('/content/drive/MyDrive/yolov8/trained_model.pt')

# Path to a training image
train_image_path = '/content/drive/MyDrive/yolov8/dataset/images/train/applsci-13-08341-g001-550_jpg.rf.64b6de9c1e0245b0380160dda69e679c.jpg'

# Run inference
results = model(train_image_path, conf=0.25)  # Lower confidence threshold for testing

# Visualize results
annotated_frame = results[0].plot()

# Display the annotated image
plt.figure(figsize=(10, 10))
plt.imshow(cv2.cvtColor(annotated_frame, cv2.COLOR_BGR2RGB))
plt.axis('off')
plt.show()

# Print detection details
for result in results:
    boxes = result.boxes
    for box in boxes:
        cls_id = int(box.cls)
        cls_name = model.names[cls_id]
        conf = box.conf
        print(f"Detected {cls_name} with confidence {conf:.2f}")
```

```python
import cv2
from ultralytics import YOLO
import matplotlib.pyplot as plt

# Load the YOLO model (make sure the model path is correct)
model = YOLO('/content/drive/MyDrive/yolov8/trained_model.pt')

# Path to the image you want to test
image_path = '/content/drive/MyDrive/yolov8/dataset/test_img_1.jpg'

# Load the image using OpenCV
img = cv2.imread(image_path)

# Run prediction on the image
results = model(img)

# Visualize the results
annotated_img = results[0].plot()  # Annotate the image with bounding boxes

# Display the image with predictions
plt.imshow(cv2.cvtColor(annotated_img, cv2.COLOR_BGR2RGB))
plt.axis('off')  # Hide axes
plt.show()

# Print the prediction results
for r in results:
    print(f"Detected classes: {r.names}")
    print(f"Bounding boxes: {r.boxes.xyxy}")
    print(f"Confidence scores: {r.boxes.conf}")
```
